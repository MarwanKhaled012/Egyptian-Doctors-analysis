```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objs as go
from wordcloud import WordCloud
```

```python
Doctors= pd.read_csv('Egyptian Doctors Data.csv')
```

```python
Doctors
```

```python
Doctors=pd.DataFrame(Doctors)
```

```python
Doctors.isna().sum()
```

```python
Doctors=Doctors.dropna(subset=['avg_rate'])
Doctors.isna().sum()
```

```python
#expand =True :: make the strings in column
#expand =False :: make the strings in list
Doctors["Source"]=Doctors["Source"].str.split(" ",n=1,expand =True)[0]
Doctors["fees"]=Doctors["fees"].str.split(" ",n=1,expand =True)[0]
Doctors["waiting_time"]=Doctors["waiting_time"].str.split(" ",n=4,expand =True)[3]
Doctors["rate_count"]=Doctors["rate_count"].str.split(" ",n=2,expand =True)[1]
Doctors["doctor_views"]=Doctors["doctor_views"].str.split(" ",n=1,expand =True)[0]
```

```python
#axis=0 for index
#axis=1 for columns
Doctors=Doctors.drop(['pages'],axis=1)
Doctors=Doctors.drop(['clinic_location'],axis=1)
Doctors=Doctors.drop(['doctor_views'],axis=1)
Doctors=Doctors.drop(['web_scraper_order'],axis=1)
Doctors=Doctors.drop(['waiting_time'],axis=1)
```

```python
Doctors=Doctors.replace(',','',regex=True)
Doctors["rate_count"]=Doctors["rate_count"].replace('','NAN',regex=True)
```

```python
Doctors["fees"]=Doctors["fees"].astype(float )
#Doctors["waiting_time"]=Doctors["waiting_time"].astype(float )
Doctors["rate_count"]=Doctors["rate_count"].astype(int )
#Doctors["doctor_views"]=Doctors["doctor_views"].astype(int)
Doctors.info()
```

```python
Doctors.info()
```

```python
Doctors
```

```python
Doctors['total_income']=Doctors['fees']*Doctors['rate_count']
Doctors
```

```python
Doctors.rename(columns={'Source': 'ID','specialization':'Specialization','fees':'Fees','avg_rate':'Avg_Rate','rate_count':'Rate_Count','total_income':'Total_Income'}, inplace=True)
```

```python
Doctors['Specialization'].value_counts()
```

```python
Doctors=Doctors.iloc[0:15,:]
Doctors
```
